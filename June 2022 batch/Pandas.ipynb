```python
data_1 = pd.DataFrame(
       {
            "key_1": ["K0", "K0", "K1", "K2"],
            "key_2": ["K0", "K1", "K0", "K1"],
            "A": ["A0", "A1", "A2", "A3"],
            "B": ["B0", "B1", "B2", "B3"],
       }
    )
data_2 = pd.DataFrame(
       {
            "key_1": ["K0", "K1", "K1", "K2"],
            "key_2": ["K0", "K0", "K0", "K0"],
            "C": ["C0", "C1", "C2", "C3"],
            "D": ["D0", "D1", "D2", "D3"],
        }
    )
```

```python
data_1
```

```python
data_2
```

```python
pd.merge(data_1,data_2, how = 'outer',on = ['key_1',  'key_2' ])
```

```python
pd.merge(data_1,data_2, how = 'inner',on = ['key_1',  'key_2' ])
```
